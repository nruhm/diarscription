```python
import whisper 
```

Expected output: Successfully installed whisper. If it fails make sure you are using venv 3.12.11 

```python
model = whisper.load_model("base") # other models: tiny, small, medium, large
result = model.transcribe("diarscription-audio-sample-a.mp3") 
print(result["text"]) # print result of transcribed text 
```

Expected output: The full audio transcription using either the base, tiny, small, medium, or large models. Please note that if you use the large model some transcriptions will be more accurate, but it will also be more aggressive towards things such as background noise, counting that as something somebody said. 

You can also use pprint to indent the results for readability
