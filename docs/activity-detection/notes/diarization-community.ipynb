PyAnnote Community enables splitting multi-speaker audio into separate files for each speaker. This notebook demonstrates using PyAnnote to separate speakers, which can be done before or after WhisperX transcription.

Prerequisites: 
- `.venv311` (Python 3.11.9)
- pyannote.audio, torch
- HF_TOKEN: Required for model access

First, import the Pipeline class from pyannote.audio to access the speaker diarization models.

```python
import pyannote.audio
```

```python
from pyannote.audio import Pipeline
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-community-1", 
    token="HF_TOKEN")
```

Process the audio file through the pipeline. This performs VAD, speaker embedding extraction, and clustering to identify who spoke when.

```python
output = pipeline("Test1.wav")
```

Iterate through the diarization output to see each speaker's segments with their start and end timestamps.

```python
for turn, speaker in output.speaker_diarization:
    print(f"{speaker} speaks between t={turn.start:.3f}s and t={turn.end:.3f}s")
```
